## Array Math

```python
x = np.array([[1,2],[3,4]], dtype=np.float64) 
y = np.array([[5,6],[7,8]], dtype=np.float64) 
```

```python
print x
print y
print(x + y) 
print(np.add(x, y)) 
```

```python
print(x - y) 
print(np.subtract(x, y)) 
```

```python
print(x * y) ##elementwise multiplication
print(np.multiply(x, y)) 
```

```python
print(x / y) 
print(np.divide(x, y)) 
```

```python
print(np.sqrt(x))
print(np.sqrt(y))
```

```python
print(np.exp(x))
```

```python
z=np.exp(x)
print(np.exp(z))
```

```python
v = np.array([9,10]) 
w = np.array([11, 12]) 
```

```python
print(v.dot(w)) 
print(np.dot(v, w)) 
```

```python
print(x.dot(v))##matrix multiplication 
print(np.dot(x, v)) 
```

```python
print(x.dot(y)) ##matrix multiplication
print(np.dot(x, y))
```

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.01)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```
